```python
columns = ['Class', 'Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash', 
           'Magnesium', 'Total phenols', 'Flavanoids', 'Nonflavanoid phenols', 
           'Proanthocyanins', 'Color intensity', 'Hue', 'OD280_OD315', 'Proline']
```

```python
df = pd.read_csv('wine.data', names=columns)
```

```python
X = df.drop('Class',axis=1)
y = df['Class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
svc_model = SVC()
```

```python
svc_model.fit(X_train,y_train)
```

```python
predictions = svc_model.predict(X_test)
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
print(classification_report(y_test,predictions))
```

```python
param_grid = {'C': [0.01, 0.1,1, 10, 100, 1000, 10000, 100000], 'gamma': [1,0.1,0.01,0.001, 0.0001, 0.00001, 0.000001, 0.0000001]} 
```

```python
grid = GridSearchCV(SVC(),param_grid,refit=True,verbose=2)
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid_predictions = grid.predict(X_test)
```

```python
print(confusion_matrix(y_test,grid_predictions))
```

```python
print(classification_report(y_test,grid_predictions))
```
